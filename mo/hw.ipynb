```python
import sweetviz as sv
import pandas as pd

# Загрузка данных (пример с CSV, можно заменить на свой источник)
data = pd.read_csv("C:/Users/ivan.dudkov/Documents/VSCodeProjects/analyze/mo/data_input/train.csv")

# 1. Анализ одного датасета
report = sv.analyze(data)
report.show_html(filepath='EDA_report.html', open_browser=True)

# 2. Сравнение двух датасетов (например, train/test)
train_data = pd.read_csv("C:/Users/ivan.dudkov/Documents/VSCodeProjects/analyze/mo/data_input/train.csv")
test_data = pd.read_csv("C:/Users/ivan.dudkov/Documents/VSCodeProjects/analyze/mo/data_input/train.csv")
comparison_report = sv.compare([train_data, "Train"], [test_data, "Test"])
comparison_report.show_html(filepath='Comparison_report.html')

# 3. Анализ с указанием целевой переменной (для задач ML)
target_report = sv.analyze(data, target_feat="SalePrice")
target_report.show_html(filepath='Target_analysis.html')

# 4. Настройка параметров анализа
custom_report = sv.analyze(
    data,
    target_feat="SalePrice",
    feat_cfg=sv.FeatureConfig(skip="id_column"),  # Пропуск столбца
    pairwise_analysis='on'  # Автоматический анализ корреляций
)
custom_report.show_html(filepath='Custom_report.html')
```

```python
from autoviz.AutoViz_Class import AutoViz_Class
import pandas as pd
import numpy as np

# Функция для предварительной обработки данных
def preprocess_data(df):
    # Заполнение пропущенных значений
    df.fillna('MISSING', inplace=True)
    
    # Преобразование строковых дат в datetime
    date_cols = [col for col in df.columns if 'date' in col.lower() or 'year' in col.lower()]
    for col in date_cols:
        try:
            df[col] = pd.to_datetime(df[col], errors='coerce')
        except:
            pass
    
    # Удаление столбцов с уникальными значениями (ID)
    cols_to_drop = [col for col in df.columns if df[col].nunique() == len(df)]
    df.drop(cols_to_drop, axis=1, inplace=True)
    
    return df

# Загрузка и предобработка данных
df = pd.read_csv("C:/Users/ivan.dudkov/Documents/VSCodeProjects/analyze/mo/data_input/train.csv")
df = preprocess_data(df)

# Инициализация AutoViz с настройками
AV = AutoViz_Class()

# Запуск AutoViz с обработкой ошибок
try:
    dft = AV.AutoViz(
        filename='',  # Не используем файл, работаем с DataFrame
        sep=",",
        depVar="SalePrice",  # целевая переменная
        dfte=df,  # передаем обработанный DataFrame
        header=0,
        verbose=2,  # более подробный вывод
        lowess=False,
        chart_format="html",  # лучше использовать html для интерактивности
        max_rows_analyzed=150000,
        max_cols_analyzed=30,
        save_plot_dir="autoviz_plots"  # сохраняем графики в отдельную папку
    )
except Exception as e:
    print(f"Произошла ошибка: {str(e)}")
    print("Попробуем альтернативный метод с ограниченным числом признаков...")
    
    # Альтернативный вариант с меньшим числом признаков
    num_cols = df.select_dtypes(include=[np.number]).columns.tolist()
    cat_cols = df.select_dtypes(include=['object']).columns.tolist()[:10]  # берем только 10 категориальных
    reduced_df = df[num_cols + cat_cols + ['SalePrice']]
    
    dft = AV.AutoViz(
        filename='',
        sep=",",
        depVar="SalePrice",
        dfte=reduced_df,
        verbose=2,
        chart_format="html"
    )

# Дополнительный анализ после завершения
print("\nАнализ завершен. Основные результаты:")
print(f"- Проанализировано строк: {len(df)}")
print(f"- Проанализировано столбцов: {len(df.columns)}")
print(f"- Типы данных:\n{df.dtypes.value_counts()}")
```
